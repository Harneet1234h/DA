```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.plot(x,y,marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt,title('Basic Line Plot')
plt.show()
```

```python
categories =['A','B','C','D']
values=[3,7,1,8]
plt.bar(categories, values, color='pink')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 25]  # Percentages
colors = ['red', 'yellow', 'orange', 'purple']


# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True)

# Title
plt.title('Favorite Fruits of 100 Students')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = x  # Linear function

# Create plot
plt.plot(x, y, 'b')  # Blue line

# Labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Show the plot
plt.show()
```
